```python
selected_genderrows = df_pdwgender[df_pdwgender['ZEIT']>'31.12.2017']
selected_genderrows
```

```python
man_rows = selected_genderrows[selected_genderrows['MERKMAL'].str.contains('Wohnberechtigte Bevölkerung - männlich')]
man_rows['WERT'] = man_rows['WERT'].astype(int)
man_rows.columns = ['ZEIT','RAUM','MERKMAL',' man_wert']
man_rows
```

```python
grouped_man = man_rows.groupby('ZEIT')
grouped_man = grouped_man.sum()
grouped_man.reset_index(level = 0,inplace = True)
grouped_man['ZEIT'] = ['2018', '2019', '2020', '2021', '2022']
grouped_man
```

```python
woman_rows = selected_genderrows[selected_genderrows['MERKMAL'] =='Wohnberechtigte Bevölkerung - weiblich']
woman_rows['WERT'] = woman_rows['WERT'].astype(int)
woman_rows.columns = ['ZEIT','RAUM','MERKMAL',' woman_wert']
woman_rows
```
